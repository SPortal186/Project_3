```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
# seaborn style
sns.set_style('whitegrid')
# Import More Dependenices 
%matplotlib inline
import matplotlib
from datetime import datetime
# warnings 
import warnings
warnings.filterwarnings("ignore")

# Dependencies and Setup
import pandas as pd
import scipy.stats as st
import time
from datetime import datetime, timedelta, date, time
```

```python
path = "Resources/CANCER_STATES.CSV"
DATA_df = pd.read_csv(path, encoding="utf-8", header=None)

DATA_df.head()
```

```python
DATA_df_2 = DATA_df.columns=("Location", "Cause", "Sex_id", "Sex", "Year", "Mx", "Upper", "Lower")
DATA_df_2 = DATA_df.drop([0,1])
DATA_df_2.reset_index(drop=True, inplace=True)
DATA_df_2 = DATA_df_2.astype({'Mx':'float', 'Upper':'float', 'Lower':'float'})
DATA_df_2['Year'] = pd.to_datetime(DATA_df_2['Year'], format='%Y')
DATA_df_main = DATA_df_2.drop(columns=["Cause", "Sex_id"])
DATA_df_main.index = DATA_df_main.index + 1
DATA_df_main = DATA_df_main.rename_axis(index="Index")
DATA_df_main 
```
